#### 1.1 Информация о пропусках 

```python
def nan_analysis(data):
    """
    Возвращает информацию о столбцах с пропусками в таблице:
    - Количество пропусков в столбце при их наличии, %; 
    - Имена столбцов без пропусков; 
    - Имена столбцов с незначительным количеством пропусков (<5%); 
    - Имена столбцов со значительным количеством пропусков (>30%);
    """
    col_names = list(data)
    for col in col_names: 
        if data[col].isna().sum() > 0:
            print('В столбце {} найдено {} пропусков = {:.2%}'.format(col, data[col].isna().sum(), data[col].isna().sum()/data.shape[0]))

    print('')
    print('Столбцы без пропусков:')
    for col in col_names: 
        if data[col].isna().sum() == 0:
            print(col)
    print('')        
    print('Пропусков < 5%, можно проигнорировать: ')
    for col in col_names: 
        if data[col].isna().sum() > 0 and data[col].isna().sum()/data.shape[0] < 0.05:
            print(col)
    print('')
    print('Пропусков > 30%: ')        
    for col in col_names: 
        if data[col].isna().sum() > 0 and data[col].isna().sum()/data.shape[0] > 0.3: 
            print(col)
```

```python
nan_analysis(df)
```

```python
print('Пропусков в user_score, включая строки с "tbd": {:.2%}'.format( 
     (df.query('user_score == "tbd"').shape[0] + df['user_score'].isna().sum())/df.shape[0]))
```

В столбцах, содержащих информацию о рейтинге и оценках, обнаружено наибольшее количество пропусков (>40%). В столбце *user_score* к количеству пропусков NaN можно прибавить количество строк, содержащих 'tbd', т.к. это значение также означает отсутствие информации об оценке. Удалять такие строки нельзя, т.к. это грозит потерей значительной части информации, поэтому в дальнейшем необходимо оценить, какую роль в исследовании будет играть информация из этих столбцов и определить стратегию обработки пропусков в них.

#### 2.1 Обработка столбца *year_of_release*

[Ранее](#1.0) было обнаружено, что проблемы с типом данных присутствуют в двух столбцах: *year_of_release* и *user_score*. 
В столбце содержится информация только о годе релиза, без полной даты, поэтому выигрыша от формата datetime мы не получим. Преобразуем тип данных в *year_of_release* в целочисленный. 

Для того, чтобы метод `astype()` отработал корректно , необходимо избавиться от пропусков в столбце. Рассчитаем средние года выпуска игр по платформам и заполним полученными значениями пропуски.  

```python
def fillna_cat(df, col_cat, col_mean): 
    """
    df - датафрейм, в котором необходимо заполнить пропуски;
    col_cat - столбец, содержащий категории, по которым будет считаться среднее;  
    col_mean - столбец, содержащий пропуски и данные для подсчета средних. 
    """
    print('Количество пропусков до обработки:', df[col_mean].isna().sum())
    cat_list = df[col_cat].unique().tolist()
    means = []
    for i in range(len(cat_list)-1): 
        condition = df[col_cat] == cat_list[i]
        mean = round(df[condition][col_mean].mean())
        df.loc[(df[col_mean].isnull()) & (condition), col_mean] = mean 
    print('Количество пропусков после обработки:', df[col_mean].isna().sum())

fillna_cat(df, 'platform', 'year_of_release')
```

```python
#преобразование типа данных в столбце year_of_release
df = df.astype({'year_of_release':'int'}, errors='ignore') 
print('Данные в столбце year_of_release после преобразования типа: ')
display(df['year_of_release'].head(2))
```

#### 2.2  Анализ и обработка столбцов *user_score, critic_score, rating*

Предположение: т.к. процент пропусков примерно одинаково высок во всех столбцах, содержащих данные о рейтинге и оценках, большинство строк, имеющих пропуск хотя бы в одном из столбцов, будет содержать пропуски в двух оставшихся. 

```python
print('Строк с пропусками в user_score:', df[df['user_score'].isna()].shape[0])
print('Строк с пропусками и "tbd" в user_score:', df[(df['user_score'].isna()) | (df['user_score']=='tbd')].shape[0])
print('Строк с пропусками в user_score, critic_score:', df[(df['user_score'].isna()) & (df['critic_score'].isna())].shape[0])
print('Строк с пропусками в user_score, critic_score, rating:', df[(df['user_score'].isna()) & (df['critic_score'].isna()) & (df['rating'].isna())].shape[0])
```

Гипотеза подтвердилась. Заметим, что наличие 'tbd' не зависит от наличия значений в столбцах *critic_score* и *rating*, но зависит от [количества продаж](#1.0.1). Проверим гипотезу о том, что отсутствие данных об оценках и рейтинге - результат низких продаж. Для этого добавим новый столбец, содержащий суммарные продажи в Европе, Японии и Северной Америке и других странах. 

```python
# добавление столбца с данными о полных продажах 
df['total_sales'] = df[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1) 

sel_df = df[(df['user_score'].isna()) & (df['critic_score'].isna()) & (df['rating'].isna())]

print('-------------Продажи для строк без рейтинга и оценок--------------')
print('Среднее -', round(sel_df['total_sales'].mean(), 2))
print('Максимальное значение -', round(sel_df['total_sales'].max(), 2))
print('')
print('-------------Продажи для строк с рейтингом и оценками-------------')
print('Среднее -', round(df['total_sales'].mean(), 2))
print('Максимальное значение -', round(df['total_sales'].max(), 2))
print('')
print('-----------Корреляция "critic_score" с другими столбцами----------')


display(df.corr()['critic_score'])
```

Действительно, средние продажи игр с отсутствующим рейтингом и оценками ниже общего среднего на 10 млн. копий. Максимальное значение продаж для игр с полной информацией выше, чем для игр без рейтинга и оценок в 2 раза. Предположение подтвердилось. 

Заполнение такого большого количества пропусков может быть оправдано при наличии сильной корреляции между параметрами, иначе подставленные значения могут сильно исказить общую картину. Предположим, что оценка критиков зависит от количества проданных копий. Однако, видим, что корреляция с *сritic_score* для всех представленных столбцов не превышает 0,25 и является недостаточной для заполнения пропусков на основании другого столбца.  

Столбцы *user_score* и *rating* не попали в матрицу корреляции, т.к. имеют нечисловой тип данных, но т.к. раньше мы выяснили, что строки с пропусками в этих столбцах совпадают в большинстве случаев, сделаем аналогичный вывод и для них - заполнение значениями на основании других столбцов невозможно. 

Для удобства обработки столбцов с оценками в будущем, заменим NaN нулями. Это соответствует логике, которой мы руководствовались - игры с низкими продажами непопулярны, поэтому не получают оценок критиков и пользователей, а, следовательно, не имеют общего рейтинга. Также заменим нулями значения "tbd", содержащиеся в *user_score*, т.к. и в этом случае игра не имеет пользовательского рейтинга. 
После замены пропусков преобразуем тип данных в *user_score* в `float`. 

```python
#замена NaN нулями 
df['critic_score'] = df['critic_score'].fillna(0)
df['user_score'] = df['user_score'].fillna(0)
print('NaN после замены:')
display(df[['critic_score', 'user_score']].isna().sum())

#замена 'tbd' нулями
df.loc[df['user_score'] == 'tbd', 'user_score'] = 0

print('Строк с "tbd" после замены:', df.query('user_score == "tbd"').shape[0])

#преобразование типа данных 
df = df.astype({'user_score':'float'}) 
```

Т.к. столбец *rating* представлен категориями, заменим NaN на категорию NR - "no rating".  

```python
df.loc[df['rating'] != df['rating'], 'rating'] = 'NR'
print('Пропусков после замены:',df['rating'].isna().sum())
```

#### 3.2. Определение жизненного срока платформ 

Определим средний срок жизни платформы. Для этого проанализируем продажи для 10 самых крупных игровых платформ по годам. 

```python
#создание сводной таблицы с количеством продаж в год для каждой платформы 
platform_sales_year = df.pivot_table(index = ['platform', 'year_of_release'], values = ['total_sales'], aggfunc=('sum'))
platform_sales_year.columns = ['sales']
#display(platform_sales_year)


#рейтинг платформ по полному количеству продаж 
df3 = platform_sales_year.groupby(level=[0])['sales'].sum().sort_values(ascending=False).reset_index()
print('------Топ-10 платформ по количеству продаж-----')
df3.head(10)
```

*Примечание:* При первичном построении графиков в данных о платформе "DS" обнаружена аномалия. Определим продолжительность жизни платформ по интервалам лет, в которых присутствуют данные о продажах.


```python
platform_list = df3['platform'].tolist()
lifespan = []

for platf in platform_list: 
    temp_df = platform_sales_year.query('platform == @platf').reset_index()
    lspn = temp_df['year_of_release'].max() - temp_df['year_of_release'].min()
    lifespan.append(lspn)

platf_lifespan = pd.DataFrame(lifespan, platform_list).reset_index()
platf_lifespan.columns = ['platform', 'lifespan']
platf_lifespan = platf_lifespan.sort_values('lifespan', ascending = False).reset_index(drop=True) #ignore_index=True)

print('------Продолжительность жизни игровых платформ------')
display(platf_lifespan.head(10))
print('Общая продолжительность жизни:', round(platf_lifespan['lifespan'].mean()), 'лет')
print('Продолжительность жизни для 10 крупнейших платформ:', round(platf_lifespan.head()['lifespan'].mean()), 'лет')
```

Примечательно, что для 10 крупнейших платформ этот срок увеличен до 15 лет - это объясняется влиянием строки PC. PC - персональные компьютеры, выбивающееся значение продолжительности жизни в этой строке объсняется универсальностью платформы, гибким развитием и собирательностью термина. 

Полученная информация о продолжительности жизни платформ позволяет определить периоод актуальных данных.
```python
def display_side_by_side(data:list, captions:list):
    """Размещает таблицы горизонтально.
        data: список таблиц Pandas.Dataframe
        captions: список заголовков таблиц
    """
    output = ""
    combined = dict(zip(captions, data))
    for caption, df in combined.items():
        output += df.style.set_table_attributes("style='display:inline'").set_caption(caption)._repr_html_()
        output += "\xa0\xa0\xa0"
    display(HTML(output)) 
```

## 5. Формирование и проверка статистических гипотез 

**Задача:** проверить гипотезы: 

- Средние пользовательские рейтинги платформ Xbox One и PC одинаковы;
- Средние пользовательские рейтинги жанров Action и Sports разные.


<a id='5.1'></a> 
#### 5.1. Гипотеза о равенстве пользовательских рейтингов платформ Xbox One и PC  

Сформулируем нулевую и альтернативную гипотезы: 
- H0 - средняя оценка пользователей для платформ одинакова; 
- Н1 - средняя оценка пользователей для платформ различается.  

Сформируем выборки и оценим их распределения. В выборках будем использовать только те данные, где пользовательская оценка не равна нулю. 

```python
#формирование выборок
x_one = data.query('platform == "XOne" and user_score > 0')['user_score']
pc = data.query('platform == "PC" and user_score > 0')['user_score']

#визуализация распределения 
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (10, 4))
fig.suptitle('Распределения оценок пользователей для платформ',
             y = 1.1,
             fontsize = 15)

sns.distplot(x_one, bins=20, ax = axes[0])
axes[0].set_title('Xbox One', pad=10)

sns.distplot(pc, bins = 20, ax = axes[1])
axes[1].set_title('PC', pad=10)
plt.show()

#параметры распределений 
print('Средняя оценка для платформы Xbox One -',round(x_one.mean(),2))
print('Средняя оценка для платформы PC -',round(pc.mean(), 2))
print('')
print('Дисперсия выборки Xbox One -',round(x_one.var(ddof = 1), 2))
print('Дисперсия выборки PC -',round(pc.var(ddof = 1), 2))
```

Распределения можно считать нормальными, на обоих распределениях можем наблюдать небольшую положительную скошенность. 

Средние оценки по платформам незначительно отличаются, разница между дисперсиями - 1,13, будем считать их неодинаковыми. 

Для проверки гипотез будет использован критерий Стьюдента- данные количественны, имеют нормальное распределение и независимы друг от друга.

Стандартные уровни значимости, применяемые при проверке статистических гипотез - 0,05; 0,01 и 0,001, при этом уровни значимости ниже 0,01 чаще всего используются в случае высокоточных измерений, или в случае, если отвержение верной гипотезы связано с серьезными рисками. 
Для текущего исследования выберем стандартный уровень значимости - 0.05. 

```python
#проверка гипотез
alpha = 0.05 
results = st.ttest_ind(x_one, pc, equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не отвергаем нулевую гипотезу") 
```

Полученное p-value говорит о 15% вероятности получить такие средние значения при условии верности нулевой гипотезы. Принимаем гипотезу о том, что средние оценки по платформам Xbox One и PC равны. 

<a id='5.2'></a> 
#### 5.2 Гипотеза о средних пользовательских рейтингах жанров Action и Sports 

Сформулируем нулевую и альтернативную гипотезы:

- H0 - средняя оценка пользователей для жанров Action и Sports одинакова;
- H1 - средняя оценка пользователей для жанров Action и Sports различается.

Сформируем выборки и оценим их распределения. Как и ранее, для формирования выборок будем использовать только оценки, не равные нулю. 

```python
#формирование выборок
action = data.query('genre == "Action" and user_score > 0')['user_score']
sports = data.query('genre == "Sports" and user_score > 0')['user_score']

#визуализация распределения 
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (10, 4))
fig.suptitle('Распределения оценок пользователей для жанров',
             y = 1.1,
             fontsize = 15)

sns.distplot(action, bins=20, ax = axes[0])
axes[0].set_title('Action', pad=10)

sns.distplot(sports, bins = 20, ax = axes[1])
axes[1].set_title('Sports', pad=10)
plt.show()

#параметры распределений 
print('Средняя оценка для жанра Action-',round(action.mean(),2))
print('Средняя оценка для жанра Sports -',round(sports.mean(), 2))
print('')
print('Дисперсия выборки Action -',round(action.var(ddof = 1), 2))
print('Дисперсия выборки Sports -',round(sports.var(ddof = 1), 2))
```

Распределения также нормальные с положительной скошенностью. 

Средние оценки и дисперсии по жанрам отличаются, разница между дисперсиями ~ 1,4, будем считать их неодинаковыми. 
Уровень значимости примем равным 0,05. 

```python
#проверка гипотез
alpha = 0.05 
results = st.ttest_ind(action, sports, equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не отвергаем нулевую гипотезу") 
```

Полученное p-value крайне мало, что говорит об очень низкой вероятности получить такие результаты при условии правдивости нулевой гипотезы. Отвергаем ее, принимаем, что средние оценки по жанрам не равны между собой. 

**Общий вывод:** проверка статистических гипотез позволила принять, что средняя оценка пользователей для жанров Action и Sports неодинакова, в то время как средние пользовательские рейтинги платформ Xbox One и PC одинаковы.
