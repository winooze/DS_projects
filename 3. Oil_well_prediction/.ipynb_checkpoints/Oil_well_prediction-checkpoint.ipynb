#### 1.2. Подготовка данных

Во всех датафреймах присутствует столбец id типа данных object, бесполезный для обучения модели. Исключим его из данных во всех трех случаях. 

```python
df1 = data1.drop('id', axis=1)
df2 = data2.drop('id', axis=1)
df3 = data3.drop('id', axis=1) 

#df1.info()
#df2.info()
#df3.info()
```

Разделим данные на обучающую и тестовую выборки в соотношении 75:25 соответственно. 

```python
def data_split(data): 
    target = data['product']
    features = data.drop('product', axis=1)
    return features, target 

f1, t1 = data_split(df1)
X1_tr, X1_val, Y1_tr, Y1_val = train_test_split(f1, t1, test_size=.25, random_state=42)

f2, t2 = data_split(df2)
X2_tr, X2_val, Y2_tr, Y2_val = train_test_split(f2, t2, test_size=.25, random_state=42)

f3, t3 = data_split(df3)
X3_tr, X3_val, Y3_tr, Y3_val = train_test_split(f3, t3, test_size=.25, random_state=42)
```

```python
namesX = [X1_tr, X2_tr, X3_tr] 
namesY = [Y1_tr, Y2_tr, Y3_tr]
for i in range(0, len(namesX)): 
    print('Размер признаков обучающей выборки ' + str(i+1) + ': {}'.format(namesX[i].shape))
print('')    
for i in range(0, len(namesY)): 
    print('Размер целевого признака обучающей выборки ' + str(i+1) + ': {}'.format(namesY[i].shape))
```

#### 1.3. Масштабирование признаков 

```python
def scaling(train, val): 
    scaler = StandardScaler() 
    scaler.fit(train)
    train_scaled = scaler.transform(train)
    val_scaled = scaler.transform(val)
    return train_scaled, val_scaled 
```

```python
X1_tr_scaled, X1_val_scaled = scaling(X1_tr, X1_val)
X2_tr_scaled, X2_val_scaled = scaling(X2_tr, X2_val)
X3_tr_scaled, X3_val_scaled = scaling(X3_tr, X3_val)
```

## 2. Обучение и проверка модели

Для оценки моделей будут использоваться метрики MSE, RMSE и R2. 

**Функции, применяемые на данном этапе исследования**

```python
def model_tuning(X_train, Y_train):
    """
    Перебор параметров сетки, возвращает лучшую комбинацию гиперпараметров для модели, оценивает 
    по метрике MSE. 
    """
    params = {'fit_intercept':[True, False], 
             'normalize': [True, False]}
    model = LinearRegression()
    model_tun = GridSearchCV(model, params, 'neg_mean_squared_error', cv=5, n_jobs=-1)
    model_tun.fit(X_train, Y_train)
    model_parameters = model_tun.best_params_
    return model_parameters 
```

```python
def predictions(X_train, Y_train, X_val): 
    '''
    Включает в себя функцию model_tuning, обучает модель на лучших гиперпараметрах сетки, 
    возвращает предсказания модели на данных X_val
    '''
    model = LinearRegression(**model_tuning(X_train, Y_train))
    model.fit(X_train, Y_train)
    predictions = model.predict(X_val)
    return predictions
```

```python
def model_metrics(Y_true, predictions): 
    '''
    Выводит метрики модели
    '''
    mse = mean_squared_error(Y_true, predictions)
    rmse = mse ** .5
    r2 = r2_score(Y_true, predictions)
    print('MSE = ', mse)
    print('RMSE = ', rmse)
    print('R2 = ', r2)
    print('Cредний запас предсказанного сырья:', np.mean(predictions))
```

```python
def predictions_scatterplots(features_valid_scaled, Y_true, predictions):
    '''
    Отстраивает реальные и предсказанные зависимости целевого признака от параметров датасета
    '''
    fig, ax = plt.subplots(1, 3, figsize = (21, 7))
    fig.suptitle('Сравнение предсказаний модели с реальными данными')
    for i in range(0, 3): 
        ax[i].scatter(features_valid_scaled[:, i], Y_true, alpha=.7, s=1, label='реальные значения')
        ax[i].scatter(features_valid_scaled[:, i], predictions, alpha=.3, s=.5, label='предсказания модели')
        ax[i].set_xlabel('f' + str(i))
        ax[i].set_ylabel('product')
        if i == 0:
            ax[i].legend(markerscale=10)    
```

#### 2.1. Регион 1

```python
predictions_1 = predictions(X1_tr_scaled, Y1_tr, X1_val_scaled)
model_metrics(Y1_val, predictions_1)
predictions_scatterplots(X1_val_scaled, Y1_val, predictions_1)
print('Реальный средний запас:', Y1_val.mean())
```

Полученные метрики и графики свидетельствуют о низком качестве предсказаний модели. Однако, в среднем значении целевого признака модель ошиблась всего лишь на 0,07.  

*Примечание:* признак f2 имеет наибольшую корреляцию с целевым параметром (0,48).  

#### 2.2. Регион 2

```python
predictions_2 = predictions(X2_tr_scaled, Y2_tr, X2_val_scaled)
model_metrics(Y2_val, predictions_2)
predictions_scatterplots(X2_val_scaled, Y2_val, predictions_2)
print('Реальный средний запас:', Y2_val.mean())
```

На данных по второму региону нефтедобычи модель отработала практически идеально. 

#### 2.3. Регион 3 

```python
predictions_3 = predictions(X3_tr_scaled, Y3_tr, X3_val_scaled)
model_metrics(Y3_val, predictions_3)
predictions_scatterplots(X3_val_scaled, Y3_val, predictions_3)
print('Реальный средний запас:', Y3_val.mean())
```

**Вывод:** лучший результат удалось получить на втором датасете, на данных 1 и 3 качество моделей значительно хуже. Однако, во всех трех случаях среднее предсказаний практически не отличается от реального среднего выборок. 

Наибольший средний запас сырья предсказан в регионе 3, наименьший - в регионе 2. 

## 3. Подготовка к расчёту прибыли

```python
def true_predicted_df(true, predicted): 
    '''
    Объединяет реальные значения с предсказанными в датафрейм,
    сортирует его по убыванию предсказанных значений.
    '''
    s1 = true 
    s2 = pd.Series(predicted, name = 'predicted')
    df = pd.concat([s1.reset_index(), s2], axis=1).drop('index', axis=1).sort_values('predicted', ascending=False)
    return df 
```

```python
#объединение данных, coxранение в переменных  
pdata1 = true_predicted_df(Y1_val, predictions_1) 
pdata2 = true_predicted_df(Y2_val, predictions_2)
pdata3 = true_predicted_df(Y3_val, predictions_3) 

display(pdata2.head(3))
```

```python
#бюджет на разработку скважин в регионе 
WELLS_BUDGET = 10000000000 
#количество разрабатываемых скважин
WELLS_AMOUNT = 200
#стоимость одной единицы продукта (тыс.баррелей)
PRODUCT_PRICE = 450000 
```

Рассчитаем достаточный объем сырья для безубыточной разработки скважины. 

```python
well_budget = WELLS_BUDGET / WELLS_AMOUNT 
print('Бюджет на разработку одной скважины: {:.0f} млн. руб.'.format(well_budget / 1000000))

product_amount = well_budget / PRODUCT_PRICE
print('Необходимое количество сырья в скважине: {:.2f}, тыс. баррелей.'.format(product_amount))
print('')
print('Средний предсказанный запас')
print('Pегион 1: {:.2f} тыс. баррелей.'.format(pdata1['predicted'].mean()))
print('Pегион 2: {:.2f} тыс. баррелей.'.format(pdata2['predicted'].mean()))
print('Pегион 3: {:.2f} тыс. баррелей.'.format(pdata3['predicted'].mean()))
```

Во всех трех регионах средний запас нефти в скважине ниже, чем требуется для безубыточной разработки скважин в регионе. 
Следует рассмотреть 200 наиболее перспективных скважин в каждом регионе оценить их среднее предсказанное количество сырья. 

```python
print('Средний предсказанный запас 200 крупнейших скважин')
print('Pегион 1: {:.2f} тыс. баррелей.'.format(pdata1['predicted'].head(200).mean()))
print('Pегион 2: {:.2f} тыс. баррелей.'.format(pdata2['predicted'].head(200).mean()))
print('Pегион 3: {:.2f} тыс. баррелей.'.format(pdata3['predicted'].head(200).mean()))
```

Средний запас нефти в перспективных скважинах во всех трех регионах достаточен для безубыточной разработки. 

## 4. Расчёт прибыли и рисков 

**Функции, применяемые на данном этапе исследования**

```python
def profit(df, wells_amount=WELLS_AMOUNT, price=PRODUCT_PRICE, wells_budget=WELLS_BUDGET):
    '''
    Возвращает реальную прибыль по предсказанным наиболее перспективным скважинам.
    wells_amount - количество наиболее перспективных скважин, по умолчанию = 200.  
    '''
    df = df.sort_values('predicted', ascending=False)
    product_total = df['product'][:wells_amount].sum()
    profit = product_total * price - wells_budget
    return profit
```

```python
state = np.random.RandomState(42)

def bootstrap(df): 
    '''
    Генерирует тысячу подвыборок размера, аналогичного исходной. Через функцию revenue находит прибыль 
    от разработки 200 наиболее перспективных скважин подвыборки. 
    Возвращает Series из 1000 значений. 
    '''
    subsample_profits = []
    for i in range(1000): 
        subsample = df.sample(500, replace=True, random_state = state)
        profit_temp = profit(subsample)
        subsample_profits.append(profit_temp)
        profits = pd.Series(subsample_profits)
    return profits 
```

```python
def interval(values, interval=.95): 
    '''
    Вычисляет и выводит границы доверительного интервала. 
    Доверительный интервал по умолчанию = 0,95.
    '''
    lower = values.quantile((1-interval)/2)
    upper = values.quantile(interval + (1-interval)/2)
    print('{:.0%} доверительный интервал: ({:.3f}, {:.3f}), млрд.руб.'
          .format(interval, lower/10**9, upper/10**9))
```

#### 4.1. Предварительная оценка прибыльности регионов 

```python
profit_1 = profit(pdata1)
profit_2 = profit(pdata2)
profit_3 = profit(pdata3)
print('Прибыль от разработки 200 наиболее перспективных по предсказаниям скважин')
print('Регион 1: {:.3f} млрд. руб.'.format(profit_1/10**9))
print('Регион 2: {:.3f} млрд. руб.'.format(profit_2/10**9))
print('Регион 3: {:.3f} млрд. руб.'.format(profit_3/10**9))
```

На данном этапе исследования наиболее перспективным для разработки является регион 1. 

#### 4.2 Оценка рисков и прибыли с использованием техники Bootstrap

**Оценка доверительных интервалов**

```python
subsample_profits_1 = bootstrap(pdata1)
subsample_profits_2 = bootstrap(pdata2)
subsample_profits_3 = bootstrap(pdata3)
```

```python
print ('\033[1m' + 'Регион 1' + '\033[0m')
interval(values= subsample_profits_1)
print('Средняя выручка: {:.3f} млрд.руб'.format(subsample_profits_1.mean()/10**9))
print('')
print ('\033[1m' + 'Регион 2' + '\033[0m')
interval(values= subsample_profits_2)
print('Средняя выручка: {:.3f} млрд.руб'.format(subsample_profits_2.mean()/10**9))
print('')
print ('\033[1m' + 'Регион 3' + '\033[0m')
interval(values= subsample_profits_3)
print('Средняя выручка: {:.3f} млрд.руб'.format(subsample_profits_3.mean()/10**9))
```

**Оценка рисков**

Оценим риски, используя полученные бутстрепом выборки с выручкой. Для каждой выборки вычислим отношение $m\over{n}$, где $m$ - количество отрицательных значений выборки , $n$ - размер всей выборки. 

```python
subsample_profits_1 = pd.DataFrame(subsample_profits_1, columns=['profit'])
subsample_profits_2 = pd.DataFrame(subsample_profits_2, columns=['profit'])
subsample_profits_3 = pd.DataFrame(subsample_profits_3, columns=['profit'])
```

```python
df_list = [subsample_profits_1, subsample_profits_2, subsample_profits_3]
for i in range(len(df_list)): 
    print('Вероятность убытков при разработке в регионе' , str(i+1), '-', df_list[i].query('profit < 0').shape[0] / df_list[i].shape[0])
```
